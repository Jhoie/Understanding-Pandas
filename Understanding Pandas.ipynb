```python
A=[1,2,3,4]
B=[5,6,7,8]
C=[9,10,11,12]
D=[13,14,15,16]
E=[17,18,19,20]
```

```python
dm=pd.DataFrame([A,B,C,D,E])
dm
```

```python
#adding labels to the rows and columns

dm=pd.DataFrame([A,B,C,D,E], ["a","b","c","d","e"], ["w","x", "y", "z"])
dm
```

```python
#add colums

dm["f"]=   dm["w"]
dm
```

```python
#deleting a row

dm.drop("a", inplace = True)
dm
```

```python
#deleting a column

dm.drop("f", axis = 1, inplace = True)
dm
```
